```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Local1.csv', encoding='ISO-8859-1')
```

```python
data.head(1)
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
Routs=data['Line'].value_counts()

plt.pie(Routs, labels=Routs.index, autopct='%1.1f%%', startangle=90)

plt.title('Local Routs')

plt.show()
```

```python
data.head()
```

```python
tracks= data['Tracks'].value_counts()

plt.pie(tracks, labels=tracks.index, autopct= '%1.1f%%' , startangle=90)

plt.title('Mumbai Stations Tracks')

plt.show()
```

```python
plt.scatter(data['Year of Opening '],data['Line'],color = 'orange')

plt.xlabel('Year of Opening ')
plt.ylabel('Station')

plt.title=('Year of Opening Station')

plt.show()
```
