# Trignometric Function 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)


fig, (ax1, ax2, ax3) = plt.subplots(3,1)

ax1.plot(x, y1, label = 'sin(x)')
ax1.set_title('Sine Wave')
ax1.legend()

ax2.plot(x, y2, label = 'cos(x)')
ax2.set_title('Cosine Waves')
ax2.legend()

ax3.plot(x, y3, label = 'tan(x)')
ax3.set_title('Tan Waves')
ax3.legend()

plt.tight_layout()

plt.savefig('Trig_Dig.png',dpi=300)

plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(10)
y=np.arange(10,20)

plt.scatter(x,y, color='green', marker='o')
plt.xlabel=('X-Axis')
plt.ylabel=('Y-Axis')
plt.title('Scatter Plot')

plt.show()
```

# Bar Diagram

```python
x=['A','B','C','D','E']
y=np.arange(10,35,5)

plt.bar(x,y, color='green')

plt.xlabel=(x)
plt.ylabel=(y)
plt.title("Bar Graph")

plt.show()
```

# Histogram

```python
data=np.random.rand(1000)

plt.hist(data, bins=30, color='gray',edgecolor='white')

#plt.xlabel('Values')
#plt.ylabel('Frequency')
plt.title('Histogram')

plt.show()
```

# Pie Chart

```python
size=[30,20,10,30,5,5]
labels=['A','B','C','D','E','F']

plt.pie(size,labels=labels, autopct='%1.1f%%',startangle=90)

plt.title('Pie Chart')

plt.show()
```
