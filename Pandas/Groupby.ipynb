# Group by

### movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)

```python
import pandas as pd
import numpy as np

movies=pd.read_csv('imdb-top-1000.csv')
movies.head()
```

```python
genres=movies.groupby('Genre')
```

```python
genres.sum()
```

```python
# Find the top 3 genre by total earning

#movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head()
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3) 

# here is second code was do perfect caus first get that columns then sum for fast execution.
```

```python
# Find the highest avg imdb genre

movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head()
```

```python
# Top rated director

movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head()
```

```python
# find the number of movies who done by the actor

movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False).head()
```

```python
len(movies.groupby('Genre')) #how many genre after groupby

# movies['Genre'].nunique()
```

# Looping

```python
# highest rated movies in there own genre
import pandas as pd

df = pd.DataFrame(columns=movies.columns)

for group,data in genres:
    df = df.append(data[data['IMDB_Rating']== data['IMDB_Rating'].max()])
```

```python
df
```

```python
# split (apply) combine
# apply -> builtin function
genres.apply(min)
```

```python
# find number of movies that starting with "A" for each group
def foo(group):
     return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
# Find ranking of each movies in the group according to imdb score

def rank_movies(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genres.apply(rank_movies)
```
